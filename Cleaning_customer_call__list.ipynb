```python
# Import pandas
import pandas as pd
```

```python
# Load the dataset
df = pd.read_excel('Customer Call List.xlsx')
df
```

```python
#remove duplicate
df = df.drop_duplicates()
df
```

```python
# removing the not necessary colunm
df = df.drop(columns='Not_Useful_Column')
df
```

```python
# removing the "/ _ ..." from the Last Name column using str.strip()
df['Last_Name'] = df['Last_Name'].str.strip('/_...')
df
```

```python
# checking the paying customer column
df['Paying Customer']
```

```python
# replace Y to Yes , N to No, and N/a to None in the Paying customer column
df['Paying Customer'] = df['Paying Customer'].replace({'Y': 'Yes', 'N': 'No','N/a': 'None'})
df
```

```python
# cleaning address column
df['Address'] = df['Address'].replace({'N/a':'None'})
df
```

```python
df['Phone_Number']
```

```python
# convert the Phone number column into the standard phone number
df['Phone_Number']
```

```python
# Import RegEx
import re

# Function to standardize phone numbers
def standardize_phone_number(phone):
    # Check if the phone number is a string
    if isinstance(phone, str):
        # Remove all non-numeric characters
        digits = re.sub(r'\D', '', phone)

        # Check if the number has the correct length (10 digits)
        if len(digits) == 10:
            return f"{digits[:3]}-{digits[3:6]}-{digits[6:]}"  # Format as XXX-XXX-XXXX
        else:
            return phone  # Return original if not 10 digits
    else:
        return None  # Return None for non-string values (e.g., NaN)
```

```python
# Applying standardize_phone_number function to the phone number column
if 'Phone_Number' in df.columns:
  df['Phone_Number'] = df['Phone_Number'].apply(standardize_phone_number)
```

```python
# Replace N/a to None
df['Phone_Number'] = df['Phone_Number'].replace({'N/a': 'None'})
df['Phone_Number']
```

```python
df['Do_Not_Contact']
```

```python
# Replace the Y to Yes, N to No
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y': 'Yes', 'N': 'No'})
df['Do_Not_Contact']
```
